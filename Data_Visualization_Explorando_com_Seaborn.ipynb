## Teste de Hipótese

**H<sup>null</sup>**
> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**
> **A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjeta.query("sobremesa == 'Sim'").porcentagem
sem_sobremesa = gorjeta.query("sobremesa == 'Não'").porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print(f'O valor do p-value é : {r.pvalue}')
```

O p-value representa a probabilidade daquela amostra ter acontecido dentro da população. Se a chance é pequena, geralmente < 0.05, representa que um evento muito raro aconteceu, então optamos por descartar a hipótese nula, e dizer que ela pode não ser verdade.

**Para aceitarmos a hipotese alternativa, sempre atribuir o p-value menor ou igual a 0.05**

**H<sup>null</sup>**
> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**
